```python
# df의 nan값 0 으로
df = df.fillna(0)
```

```python
def to_meta_data(df):
    meta_data = []
    for col in df.columns:
        meta = {
            'metadata': col,
            'page_content': df[col].iloc[:]
        }
        meta_data.append(meta)
    return meta_data


meta_data = to_meta_data(df)
meta_data
```

```python
import json

contents = []
meta = []
for item in meta_data:
    md = item.get('metadata')
    pc = item.get('page_content', '')
    # metadata 가 dict 이면 JSON 문자열화, 아니면 그냥 str()
    if isinstance(md, dict):
        md_str = json.dumps(md, ensure_ascii=False)
    else:
        md_str = str(md)
    contents.append(f"{md_str}은 {pc}입니다")

    meta.append({
        'metadata': md,
        'page_content': pc
    })

print(meta[0])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=0,
    length_function=len
)

docs = splitter.split_documents(contents)
```

```python
# df.columns == ['개발', '경영비즈니스', ...]

def to_document(df):
    documents = []
    for col in df.columns:
        # ① 문자열로 변환
        # 1) 공백으로 연결
        text = " ".join(f"{idx}: {df.at[idx, col]}" for idx in df.index)

        # ② metadata 도 dict
        meta = {"category": col}

        documents.append(Document(page_content=text, metadata=meta))
    return documents
```
