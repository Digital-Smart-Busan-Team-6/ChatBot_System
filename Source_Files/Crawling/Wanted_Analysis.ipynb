# 직군별 기술스택의 개수 Top x

```python
import pandas as pd
import numpy as np


def top3_stack_per_job(detail_df: pd.DataFrame, tech_col: str = "기술 스택",
                       job_col: str = "직군", top_n: int = 3) -> dict:
    """
    직군별 상위 N개 기술 스택 + 개수를 JSON-용 딕셔너리로 반환
    --------------------------------------------------------
    반환 예:
    {
      "직군별 기술 스택 TOP3": {
        "백엔드":     {"Python": 128, "Java": 97,  "Spring": 91},
        "프론트엔드": {"React": 110, "JavaScript": 88, "TypeScript": 49},
        ...
      }
    }
    """

    # 1) 기술 스택 리스트를 행으로 펼치기
    df = detail_df[[job_col, tech_col]].explode(tech_col, ignore_index=True)

    # 2) 문자열화 + 공백 제거
    df[tech_col] = (
        df[tech_col]
        .fillna("")  # NaN → 빈문자
        .astype(str)
        .str.strip()
    )

    # 3) ***빈 문자열, 'nan', 직군 이름 자체 제거***
    mask_valid = (
            (df[tech_col] != "") &
            (df[tech_col].str.lower() != "nan") &
            (df[tech_col] != df[job_col])  # '백엔드'가 기술 스택에 섞이는 경우 방지
    )
    df = df[mask_valid]

    # 4) (직군, 기술) 빈도
    counts = (
        df.groupby([job_col, tech_col])
        .size()
        .rename("cnt")  # MultiIndex Series: (직군, 기술) → cnt
    )

    # 5) 직군별 상위 N 추출
    result = {}
    for job, sub in counts.groupby(level=0):
        top = (
            sub.sort_values(ascending=False)
            .head(top_n)
            .droplevel(0)  # 인덱스에서 직군 부분 제거 → 기술만 남김
        )
        result[job] = {tech: int(cnt) for tech, cnt in top.items()}

    return {f"직군별 기술 스택 TOP{top_n}": result}


# ── 사용 예 ────────────────────────────────────────────
top3_json = top3_stack_per_job(detail_df, top_n=5)

original = getOriginalJson(FILE_PATH)
merged = mergeJsonDicts(original, top3_json)
saveData(merged, FILE_PATH)
```

# 직군별 신입 지원 가능 여부

```python
# 1) (직군, 신입 여부)별 갯수 집계 → DataFrame
newbie_df = (
    detail_df.groupby(['직군', '신입 지원 가능 여부'])
             .size()                  # MultiIndex Series
             .unstack(fill_value=0)   # 칼럼: False / True
)

newbie_true = newbie_df.get(True, pd.Series(dtype=int))

# 3) Series → {직군: 개수} 딕셔너리
category_newbie_count = {
    '직군별 신입 지원 가능 여부': newbie_true.astype(int).to_dict()
}

# 결과 예시
# {
#   '직군별 신입 지원 가능 여부': {
#       '백엔드': 120,
#       '프론트엔드': 98,
#       '데이터': 45,
#       ...
#   }
# }
original_json = getOriginalJson(FILE_PATH)
merge = mergeJsonDicts(original_json, category_newbie_count)
saveData(merge, FILE_PATH)
```
