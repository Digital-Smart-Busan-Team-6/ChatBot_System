```python
titleInfo = body.find_all('div', class_ = "sc-923c3317-0 jIWuEG")
h1_titles = []
for info in titleInfo:
    h1_tag = info.find('h1')
    if h1_tag:  # <h1> 태그가 존재하는 경우에만 추가
        h1_titles.append(h1_tag.text)

print(h1_titles)
```

```python
company = []
for info in titleInfo:
    company_tag = info.find('a')
    if company_tag:  # <h1> 태그가 존재하는 경우에만 추가
        company.append(company_tag.text)
print(company)
```

```python
pride = []
for info in titleInfo:
    pride_tags = info.find_all('li')
    for pride_tag in pride_tags:
        pride.append(pride_tag.text)
print(pride)
```
