```python
from Data_Analysis import *
from Jumpit_Crawling_Functions import *
import pandas as pd
```

```python
positions = getJumpitPositions()

ids = getIDs(positions)
titles = getTitles(positions)
jobCategories = getJobCategories(positions)
companyNames = getCompanyNames(positions)
minCareer = getMinCareer(positions)
maxCareer = getMaxCareer(positions)
techStacks = getTechStacks(positions)
jobLocations = getJobLocations(positions)

mainDataFrame = pd.DataFrame({
    'id' : ids,
    'jobCategory': jobCategories,
    'techStack': techStacks,
    'minCareer': minCareer,
    'maxCareer': maxCareer,
    'jobLocation': jobLocations,
})

mainDataFrame.set_index('id', inplace=True)
```

```python
data = crawlingDetailPage()
detailDataFrame = pd.DataFrame(data)
detailDataFrame.set_index('id', inplace=True)
detailDataFrame.drop(columns=['jobLocation', 'career','techStack'], axis=1, inplace=True)
```

```python
mergedDataFrame = combineDataFrames(mainDataFrame, detailDataFrame)
```

```python
improvedDataFrame = toImproveDataFrame(mergedDataFrame)
```

```python
# education의 대학졸업2,3년 이상을 모두 대학졸업2년 이상 으로 변경
improvedDataFrame['education'] = improvedDataFrame['education'].replace('대학졸업2,3년 이상', '대학졸업2년 이상')
```

```python
# 중복 제거
improvedDataFrame.drop_duplicates(subset=['title'], keep='first', inplace=True)
```
